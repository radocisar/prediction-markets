```python
import pandas as pd
import json
from IPython.display import display
```

```python
# events = pd.read_json('events.json')
# events = pd.read_json('fed.json')
with open("fed.json", "r") as f:
    # data = json.load(f)
    events = pd.json_normalize(json.load(f), record_path=["markets"], record_prefix="markets.", meta=["ticker", "slug", "title"])
events.head()
```

```python
events.info()
```

```python
events.melt(id_vars="id", var_name='columns', value_name='values')
```

```python
with pd.option_context("display.max_colwidth", None):
    display(events.melt(id_vars="id", var_name='columns', value_name='values'))
```

```python
# Series
with pd.option_context("display.max_colwidth", None):
    with pd.option_context("display.max_rows", None):
        display(pd.json_normalize(events["series"].apply(pd.Series)[0]).melt(id_vars="id", var_name="columns", value_name="values"))
```
