```python
import pandas as pd
import json
from IPython.display import display
```

```python
# events = pd.read_json('events.json')
# events = pd.read_json('fed.json')
with open("fed.json", "r") as f:
    # data = json.load(f)
    events = pd.json_normalize(
        json.load(f),
        record_path=["markets"],
        record_prefix="markets.",
        meta=[
            "id",
            "ticker",
            "slug",
            "title",
            "description",
            "resolutionSource",
            "startDate",
            "creationDate",
            "endDate",
            "image",
            "icon",
            "active",
            "closed",
            "archived",
            "new",
            "featured",
            "restricted",
            "liquidity",
            "volume",
            "openInterest",
            "createdAt",
            "updatedAt",
            "competitive",
            "volume24hr",
            "volume1wk",
            "volume1mo",
            "volume1yr",
            "enableOrderBook",
            "liquidityClob",
            "negRisk",
            "negRiskMarketID",
            "commentCount",
            "series",
            "tags",
            "cyom",
            "showAllOutcomes",
            "showMarketImages",
            "enableNegRisk",
            "automaticallyActive",
            "seriesSlug",
            "gmpChartMode",
            "negRiskAugmented",
            "featuredOrder",
            "estimateValue",
            "cumulativeMarkets",
            "pendingDeployment",
            "deploying",
            "deployingTimestamp",
            "requiresTranslation",
        ],
    )
events.head()
```

```python
events.melt(id_vars="id", var_name="columns", value_name="values")
```

```python
with pd.option_context("display.max_colwidth", None):
    with pd.option_context("display.max_rows", None):
        display(events.melt(id_vars="id", var_name="columns", value_name="values"))
```
